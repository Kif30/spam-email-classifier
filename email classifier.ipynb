#***Spam email detection***



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('email.csv')
```

```python
df
```

```python
df.loc[:,'class']=df.loc[:,'Category'].map({'ham':1,'spam':0})
```

```python
df.drop(columns=['Category'],axis=1,inplace=True)
```

```python
df.loc[:,'class'].unique()
```

```python
df.dropna(subset=['class'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.loc[5:'message']
```

```python
import re
import string
```

```python
def clean_text (text):
  text= text.lower()
  # remove html tags
  text = re.sub(r'<.*?>',"",text)
  # remove urls
  text = re.sub(r'http\S+|www\S+','',text)
  # remove emojis
  text = re.sub(r'[^x00-\x7F]+',' ',text)
  # special characters
  text = re.sub(r'[^a-z0-9\s]','',text)
  # white space
  text = re.sub(r'\s+',' ',text).strip()

  return text
```

```python
df.loc[:,'cleaned_text']=df.loc[:,'Message'].apply(clean_text)
```

```python
df
```

```python
x=df['cleaned_text']
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#count vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_new_train=cv.fit_transform(x_train)
x_new_test=cv.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_new_train,y_train)
y_pred=model.predict(x_new_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,f1_score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

#predictions

```python
msg=input("enter the message")
test=cv.transform([msg])
pred=model.predict(test)
if pred[0]==0:
 print('spam')
else:
  print("ham")
```
